From the above graphs, we can see that there are some outliers present in capacity_mw,longitude and source columns. We need to remove outliers

```python
 #Removing Outliers using zscore and IQR method
    
from scipy.stats import zscore
```

```python
out_features = df[['capacity_mw','longitude','source','geolocation_source']]
z =np.abs(zscore(out_features))
z
```

```python
np.where(z>3)
```

```python
df1 = df[(z<3).all(axis=1)]
```

```python
df1.shape
```

```python
print("First DataFrame -", df.shape[0])
print("New DataFrame -", df1.shape[0])
```

```python
print (" Data Loss percentage after removing outliers with z-score method -", ((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

```python
#Lets check Data Loss percentage using IQR method

#1st Quantile
Q1 = out_features.quantile(0.25)

#3RD Quantile
Q3 = out_features.quantile(0.75)

#IQR
IQR = Q3 -Q1
df2 = df[~((df<(Q1 -1.5 *IQR))|(df>(Q3 + 1.5 * IQR))).any(axis =1)]
```

```python
df2.shape
```

```python
print (" Data Loss percentage after removing outliers with IQR method -", ((df.shape[0]-df2.shape[0])/df.shape[0])*100)
```

Here Data loss percentage using IQR method is more than 44% which is not in acceptable range and zscore is only 5.6% which is affordable.

```python
#Splitting the data into x and y

x = df.iloc[:, 1:8].values
y = df.iloc[: ,0].values
```

```python
#Splitting the data into training and testing

from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size =0.20, random_state =0)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
#Applying ML models 
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
RFC = RandomForestClassifier()
RFC.fit(x_train,y_train)
print('calculating the error')
PredRFC = RFC. predict(x_test)
print(accuracy_score(y_test,PredRFC))
print(confusion_matrix(y_test, PredRFC))
print(classification_report(y_test, PredRFC))
```

```python
DTC = DecisionTreeClassifier(criterion='entropy', random_state =0)
DTC.fit(x_train,y_train)
print('calculating the error')
PredDTC = DTC. predict(x_test)
print(accuracy_score(y_test,PredDTC))
print(confusion_matrix(y_test, PredDTC))
print(classification_report(y_test, PredDTC))
```

```python
LR = LogisticRegression()
LR.fit(x_train,y_train)
print('calculating the error')
PredLR = LR. predict(x_test)
print(accuracy_score(y_test,PredLR))
print(confusion_matrix(y_test, PredLR))
print(classification_report(y_test, PredLR))
```

```python
GBR = GradientBoostingClassifier()
GBR.fit(x_train,y_train)
print('calculating the error')
PredGBR = GBR. predict(x_test)
print(accuracy_score(y_test,PredGBR))
print(confusion_matrix(y_test, PredGBR))
print(classification_report(y_test, PredGBR))
```

```python
ADB = AdaBoostClassifier()
ADB.fit(x_train,y_train)
print('calculating the error')
PredADB = ADB. predict(x_test)
print(accuracy_score(y_test,PredADB))
print(confusion_matrix(y_test, PredADB))
print(classification_report(y_test, PredADB))
```

```python
#Accuracy score of most of the models are 96.5%. Lets consider RFC for prediction


x_predict = list(RFC.predict(x_test))
Predicted_df = {'Predicted_values': x_predict, 'Actual_values': y_test}
pd.DataFrame(Predicted_df).head(10)
```
