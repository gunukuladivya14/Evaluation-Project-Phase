```python
#Checking for Outliers

column = []
for i in df.dtypes.index:
    if df.dtypes[i]!= 'object':
        column.append(i)
print("Column", column)
        
```

```python
plt.figure(figsize = (20,12), facecolor ='white')
plotnumber = 1
for col in column:
    if plotnumber<=17:
        ax = plt.subplot(6,17, plotnumber)
        sns.boxplot(df[col])
        plt.xlabel(col, fontsize =20)
        plt.yticks(rotation=0,fontsize=15)
    plotnumber+=1
plt.tight_layout()
```

In the above plots, we can clearly see that there are some outliers present in few Columns (R, ERA, SHO, SV and E).

```python
 #Removing Outliers using zscore and IQR method

from scipy.stats import zscore
```

```python
out_features = df[['R','ERA','SHO','SV','E']]
z =np.abs(zscore(out_features))
z
```

```python
np.where(z>3)
```

```python
df1 = df[(z<3).all(axis=1)]
```

```python
df1.shape
```

```python
print("First DataFrame -", df.shape[0])
print("New DataFrame -", df1.shape[0])
```

```python
print("Data Loss percentage -",((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

```python
#1st Quantile
Q1 = out_features.quantile(0.25)

#3RD Quantile
Q3 = out_features.quantile(0.75)

#IQR
IQR = Q3 -Q1
df2 = df[~((df<(Q1 -1.5 *IQR))|(df>(Q3 + 1.5 * IQR))).any(axis =1)]
```

```python
df2.shape
```

```python
print (" Data Loss percentage after removing outliers with IQR method -", ((df.shape[0]-df2.shape[0])/df.shape[0])*100)
```

From the above two methods, Data Loss percentage in Z-score method is 3% and IQR method is 33%. Data Loss percentage in z score is acceptable as it is less than 10%. So, I will go ahead with that data loss

```python
df = df.drop(['H','AB'], axis =1)
df
```

```python
#Splitting the data into x and y

x = df.iloc[:, 1:15].values
y = df.iloc[: ,0].values
```

```python
x
```

```python
y
```

```python
#Splitting the data into training and testing

from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size =0.20, random_state =0)
```

```python
#Scaling the data using Standard Scaler

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
#Applying ML Models
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(n_estimators =15, criterion='squared_error', random_state =None)
RFR.fit(x_train,y_train)
predRFR = RFR.predict(x_test)
print("RandomForestRegressor training Accuracy: ", RFR.score(x_train, y_train))
```

```python

from sklearn.ensemble import GradientBoostingRegressor
GBR = GradientBoostingRegressor(n_estimators=20, random_state =None)
GBR.fit(x_train,y_train)
predGBR = GBR.predict(x_test)
print("GradientBoostingRegressor training Accuracy: ", GBR.score(x_train, y_train))
```

```python

from sklearn.ensemble import AdaBoostRegressor
ABR = AdaBoostRegressor(estimator =None, random_state =None)
ABR.fit(x_train,y_train)
predABR = ABR.predict(x_test)
print("AdaBoostRegressor training Accuracy: ", ABR.score(x_train, y_train))
```

Accuracy score of Adaboost Regressor is high compared to other models.

```python
ABR = AdaBoostRegressor()
ABR.fit(x_train,y_train)
```

```python
predABR = ABR.predict(x_test)
predABR
```
